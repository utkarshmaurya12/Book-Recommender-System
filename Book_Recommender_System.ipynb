```python
ratings_with_book = ratings.merge(books,on='ISBN')
ratings_with_book
```

```python
no_ratings = ratings_with_book.groupby('Book-Title').count()['Book-Rating'].reset_index()
no_ratings
```

```python
no_ratings.rename(columns={'Book-Rating': 'no_ratings'},inplace=True)
```

```python
no_ratings
```

```python
avg_ratings = ratings_with_book.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratings.rename(columns={'Book-Rating': 'avg_ratings'},inplace=True)
avg_ratings
```

```python
popularity = no_ratings.merge(avg_ratings,on='Book-Title')
popularity
```

```python
popularity = popularity[popularity['no_ratings']>=150]
popularity
```

```python
popularity=popularity.sort_values('avg_ratings',ascending=False).head(50)
popularity
```

```python
popularity = popularity.merge(books,on='Book-Title')
popularity
```

```python
popularity = popularity.drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','no_ratings','avg_ratings']]
popularity
```

```python
x = ratings_with_book.groupby('User-ID').count()['Book-Rating'] > 200
x[x].index
```

```python
ratings_taken = x[x].index
```

```python
ratings_with_book['User-ID'].isin(ratings_taken)
```

```python
filtered_rating = ratings_with_book[ratings_with_book['User-ID'].isin(ratings_taken)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
```

```python
famous_books = y[y].index
```

```python
filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating.drop_duplicates()
```

```python
final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
df = final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
df.fillna(0, inplace=True)
```

```python
df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(df)
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(df.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == df.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('You Belong To Me')
```
